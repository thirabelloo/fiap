```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Visualizando a base de dados disponibilizada
dataset = "../data/Banco de Dados Empresa - Base.csv"
df = pd.read_csv(dataset)
df.head()
```

```python
# Verificando a quantidade de linhas e colunas do DataFrame
df.shape
```

```python
# Visualizando informacoes do DataFrame
df.info()
```

```python
# Obtendo dados estatisticos do Dataframe
df.describe()
```

```python
# Obtendo dados estatistico da coloca Idade
df["Idade"].describe()
```

```python
# Obtendo a media da coluna 'Idade'
df["Idade"].mean()
```

```python
# Obtendo a mediana da coluna 'Idade'
df["Idade"].median()
```

```python
# Plotando o grafico tipo box da coluna 'Idade'
df["Idade"].plot(kind="box", vert=False, figsize=(14, 6))
```

```python
# Plotando o grafico tipo density da coluna 'Idade'
df["Idade"].plot(kind="density", figsize=(14, 6))
```

```python
# Plotando o grafico tipo density da coluna 'Idade' com media e mediana
ax = df["Idade"].plot(kind="density", figsize=(14, 6))
ax.axvline(df["Idade"].mean(), color="red")
ax.axvline(df["Idade"].median(), color="green")
```

```python
# Plotando o grafico tipo hist da coluna 'Idade'
ax = df["Idade"].plot(kind="hist", figsize=(14, 6))
ax.set_ylabel("Funcionarios")
ax.set_xlabel("Idade")
```

```python
# Visualizando o DataFrame
df.head()
```

```python
# Calculando a quantidade da coluna 'CampoEducacao'
df["CampoEducacao"].value_counts()
```

```python
# Plotando um grafico tipo pie da coluna 'CampoEducacao'
df["CampoEducacao"].value_counts().plot(kind="pie", figsize=(6, 6))
```

```python
# Plotando um grafico tipo bar da coluna 'CampoEducacao'
ax = df["CampoEducacao"].value_counts().plot(kind="bar", figsize=(14, 6))
ax.set_ylabel("Campo de Educação")
```

```python
# Realizando o calculo de correlacao
df_numerico = df.select_dtypes(include=["float64", "int64"])
correlacao = df_numerico.corr()
correlacao
```

```python
# Plotando o grafico
fig = plt.figure(figsize=(8, 8))
plt.matshow(correlacao, cmap="RdBu", fignum=fig.number)
plt.xticks(range(len(correlacao.columns)), correlacao.columns, rotation="vertical")
plt.yticks(range(len(correlacao.columns)), correlacao.columns)
```

```python
# Plotando um grafico tipo scatter
df.plot(kind="scatter", x="AnosMesmoGestor", y="AnosEmpresa", figsize=(14, 6))
```

```python
# Utilizando o boxplot para plotar um grafico
ax = df[["NivelnoTrabalho", "Renda Mensal"]].boxplot(
    by="NivelnoTrabalho", figsize=(14, 6)
)
ax.set_ylabel("Renda Mensal")
```

```python
boxplot_colunas = ["Idade", "Diária", "DistanciadeCasa"]
df[boxplot_colunas].plot(kind="box", subplots=True, figsize=(14, 8))
```
